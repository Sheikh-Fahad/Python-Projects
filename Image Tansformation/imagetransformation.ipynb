Reflection

```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow



img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
rows, cols = img.shape
M = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])
reflected_img = cv.warpPerspective(img, M,
                                   (int(cols),
                                    int(rows)))
cv2_imshow(img)
cv2_imshow(reflected_img)


```

Rotation

```python



img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
rows, cols = img.shape
M = np.float32([[1,  0, 0], [0, -1, rows], [0,  0, 1]])

img_rotation = cv.warpAffine(img,
                             cv.getRotationMatrix2D((cols/2, rows/2),
                                                    30, 0.6),
                             (cols, rows))
cv2_imshow(img)
cv2_imshow(img_rotation)
```

Image Scaling

```python

import numpy as np
import cv2 as cv
img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
rows, cols = img.shape
img_shrinked = cv.resize(img, (250, 200),
                         interpolation=cv.INTER_AREA)
cv2_imshow(img)
cv2_imshow(img_shrinked)



img_enlarged = cv.resize(img_shrinked, None,
                         fx=1.5, fy=1.5,
                         interpolation=cv.INTER_CUBIC)
cv2_imshow(img_enlarged)
cv2_imshow(img)




```

Image Cropping

```python


img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
cropped_img = img[100:300, 100:300]
cv2_imshow(img)
cv2_imshow(cropped_img)


```

Image Shearing in X-Axis

```python


img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
rows, cols = img.shape
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
cv2_imshow(img)
cv2_imshow(sheared_img)
```

Image Shearing in Y-Axis

```python


img = cv.imread('drive/MyDrive/AlienX.jpeg', 0)
rows, cols = img.shape
M = np.float32([[1,   0, 0], [0.5, 1, 0], [0,   0, 1]])
sheared_img = cv.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))

cv2_imshow(img)
cv2_imshow(sheared_img)
```

 Applies Gaussian blur to the image.

```python


img_blurred = cv.GaussianBlur(img, (5, 5), 0)
cv2_imshow(img_blurred)
```

Performs histogram equalization on the image.

```python


img_equalized = cv.equalizeHist(img)
cv2_imshow(img_equalized)
```

Applies binary thresholding.

```python


_, img_thresh = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
cv2_imshow(img_thresh)
```

 Applies Canny edge detection to the image.

```python


edges = cv.Canny(img, 100, 200)
cv2_imshow(edges)
```

Image warped

```python
# Define four corners of the region of interest.

pts1 = np.float32([[50, 50], [200, 50], [50, 200], [200, 200]])

# Define how the points will be positioned after the transformation.

pts2 = np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])

M_perspective = cv.getPerspectiveTransform(pts1, pts2)
img_warped = cv.warpPerspective(img, M_perspective, (300, 300))
cv2_imshow(img_warped)
```

Add Gaussian noise to the image.

```python


mean = 0
std_dev = 25
noisy_img = img + np.random.normal(mean, std_dev, img.shape)
cv2_imshow(noisy_img)
```

Morphological Operations

```python
kernel = np.ones((5, 5), np.uint8)
img_eroded = cv.erode(img, kernel, iterations=1)
img_dilated = cv.dilate(img, kernel, iterations=1)
cv2_imshow(img_eroded)
cv2_imshow(img_dilated)
```
